```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = ' '.join([word for word in word_tokenize(text) if word not in stopwords.words('english')])  # Remove stopwords
    return text

preprocessed_documents = [preprocess_text(doc) for doc in documents]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(preprocessed_documents)

num_clusters = 5  # Number of clusters (you can adjust this)
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

cluster_labels = kmeans.labels_

df = pd.DataFrame({'Document': documents, 'Cluster': cluster_labels})

for i in range(num_clusters):
    print(f"Cluster {i}:")
    print(df[df['Cluster'] == i]['Document'])
    print()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud


tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix = tfidf_vectorizer.fit_transform(preprocessed_documents)


feature_names = tfidf_vectorizer.get_feature_names_out()

```

```python
for i in range(num_clusters):
    print(f"Cluster {i}:")
    cluster_indices = df[df['Cluster'] == i].index
    cluster_keywords = []
    for idx in cluster_indices:
        document_keywords = []
        for word_idx in tfidf_matrix[idx].nonzero()[1]:
            document_keywords.append(feature_names[word_idx])
        cluster_keywords.extend(document_keywords)


    keyword_counter = Counter(cluster_keywords)


    plt.figure(figsize=(10, 6))


    if len(keyword_counter) > 0:
        keywords, frequencies = zip(*keyword_counter.most_common(10))


        plt.bar(keywords, frequencies)
        plt.title(f"Cluster {i} - Top Keywords")
        plt.xlabel("Keyword")
        plt.ylabel("Frequency")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()


        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(keyword_counter))
        plt.figure(figsize=(10, 6))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.title(f"Cluster {i} - Keyword Word Cloud")
        plt.axis('off')
        plt.show()
    else:
        print("No keywords found for this cluster.")
```

```python
import matplotlib.pyplot as plt


cluster_counts = df['Cluster'].value_counts().sort_index()


plt.figure(figsize=(8, 6))
plt.bar(cluster_counts.index, cluster_counts.values, color='skyblue')
plt.title('Document Count in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Document Count')
plt.xticks(cluster_counts.index)
plt.show()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from wordcloud import WordCloud


lda_model = LatentDirichletAllocation(n_components=5, random_state=42)
lda_model.fit(tfidf_matrix)


topic_keywords = {}
for topic_idx, topic in enumerate(lda_model.components_):
    keywords_idx = topic.argsort()[:-11:-1]
    topic_keywords[f"Topic {topic_idx+1}"] = [feature_names[i] for i in keywords_idx]


plt.figure(figsize=(15, 10))
for i, (topic, keywords) in enumerate(topic_keywords.items(), start=1):
    plt.subplot(2, 3, i)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(keywords))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(topic)
    plt.axis('off')
plt.tight_layout()
plt.show()
```
